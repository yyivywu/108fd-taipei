```python
import requests
```

```python
# 使用 requests.get 函式讀取指定的網頁內容

url = "https://taqm.epa.gov.tw/pm25/tw/PM25A.aspx?area=1"
data = requests.get(url)
```

```python
# 伺服器回應的狀態碼, 如果顯示 200 就代表讀取 ok

print (data.status_code)
```

```python
# 用判斷式判斷是否讀取 ok

if data.status_code == requests.codes.ok:
    print ("網頁讀取 ok")
```

```python
# 輸出網頁 HTML 原始碼

print(data.text)
```

```python
# 查詢網頁編碼

print(data.encoding)
```

```python
# 輸出網頁 header 內容

print(data.headers)
```

```python
# 設定 url 的查詢參數
my_params = {'area': '4'}

# 將查詢參數加入 GET 請求中
url = 'https://taqm.epa.gov.tw/pm25/tw/PM25A.aspx'
data = requests.get(url, params = my_params)
```

```python
# 輸出擷取的 URL
print(data.url)
```

```python
# 尋找網頁內容中的特定字串

if '西屯' in data.text:
    print("找到!")
else:
    print("沒找到!")
```

```python
# 分割網頁內容為串列

datalist = data.text.splitlines()

for row in datalist:
    print(row)
```

```python
# 計算特定字串出現在列的次數

n = 0
for row in datalist:
    if '空氣' in row:
        n += 1

print('找到空氣 %d 次' % n)
#print ('找到空氣 {} 次'.format(n))
```
