# Pandas 模組－繪圖應用

參數說明：
* kind：圖形種類 line 線形圖、hist 直方圖、scatter 散點圖、bar 直向長條圖、barh 橫向長條圖、pie 圓餅圖
* title：繪製圖形標題
* legend：是否顯示圖例說明
* grid：是否顯示格線
* xlim：X 軸的刻度範圍
* ylim：y 軸的刻度範圍
* xticks：x 軸的刻度值
* yticks：y 軸的刻度值
* fontsize：x,y 軸刻度的字體大小
* figsize：繪製圖形的長度及寬度，資料型態為 tuple (width, height)，吋 (inch) 為單位

```python
import pandas as pd
%matplotlib inline
```

```python
datas = [[65,92,78,88], [90,91,60,55], [80,40,99,78],[88,70,87,99]]
indexs = ["王大一", "張美麗", "林田山", "李丁一"]
columns = ["國文", "數學", "英文", "自然"]

df = pd.DataFrame(datas, columns=columns, index=indexs)

#使用 plot 繪圖
df.plot(kind="line", title="三年甲班成績", legend=True, fontsize=12, figsize=(9,6))  #線形圖
```

```python
df.plot(kind="bar", title="三年甲班成績", legend=True, fontsize=10, figsize=(9,6), grid=True)
```
