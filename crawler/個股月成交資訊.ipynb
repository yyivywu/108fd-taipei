# 讀取個股月成交資訊

網站：[TWSE台灣證券交易所](https://www.twse.com.tw/zh/page/trading/exchange/FMSRFK.html)

```python
import requests

url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20190930&stockNo=2330'
stock = requests.get(url)
#stock.encoding= 'utf-8'
print(stock.text)

# fout = open('stock.html', 'w', encoding='utf-8')
# fout.write(stock.text)
# fout.close
```

# 輸入要查詢的個股代號，動態爬取資料

```python
import requests

sid = input('請輸入股票代碼:')
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20190101&stockNo='+sid
stock = requests.get(url)
print(stock.text)
```

# 使用參數傳遞寫法

```python
import requests

syear = input('請輸入查詢年度(格式：20190101)：')
sid = input('請輸入查詢個股代號：')

# 查詢參數
stock_params = {'response': 'html', 'date': syear, 'stockNo': sid}

# 將查詢參數加入 GET 請求中
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK'
stock = requests.get(url, params = stock_params)

print(stock.text)
```
